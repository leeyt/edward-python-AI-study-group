## Simple Plot

Please note, this script is for python3+.
If you are using python2+, please modify it accordingly.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(-1, 1, 50)
y = 2*x + 1

plt.plot(x, y)
plt.show()
```

```python
y = x**2

plt.plot(x, y)
plt.show()
```
