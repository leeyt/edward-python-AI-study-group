```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2
```

```python
plt.figure()
plt.plot(x, y1)
```

```python
plt.figure(num=3, figsize=(8, 5),)
plt.plot(x, y2)

# plot the second curve in this figure with certain parameters
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.show()
```
